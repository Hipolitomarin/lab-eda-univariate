```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)
```

```python
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
df.head()
```

```python
category_counts = df['category'].value_counts()
category_counts.head(10)
```

```python
top_5_categories = category_counts.head(5)
top_5_categories
```

```python
category_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 category by number of products')
plt.xlabel('Category')
plt.ylabel('Number of products')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_5_categories.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Top 5 catgeories')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
```

```python
mean_price, median_price, mode_price
```

```python
price_variance = df['price'].var()
price_std = df['price'].std()
price_range = df['price'].max()-df['price'].min()
price_itq = df['price'].quantile(0.75)-df['price'].quantile(0.25)
price_variance, price_std, price_range, price_itq 
```

```python
df[df['price'] < 500]['price'].hist(bins=50)
plt.title('Distribution of prices (< £500)')
plt.xlabel('Prices')
plt.ylabel('Count of products')
plt.show()
```

```python
sns.boxplot(x=df['price'])
plt.title('Boxplot of price of products')
plt.xlabel('Price')
plt.show()
```

```python
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]

mean_rating, median_rating, mode_rating
```

```python
rating_var = df['stars'].var()
rating_std = df['stars'].std()
rating_iqr = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

rating_var, rating_std, rating_iqr
```

```python
rating_skew = df['stars'].skew()
rating_kurt = df['stars'].kurtosis()

rating_skew, rating_kurt
```

```python
df['stars'].hist(bins=10, color='orange')
plt.title('Distribution of ratings')
plt.xlabel('Rating')
plt.ylabel('Quantity of products')
plt.show()
```
